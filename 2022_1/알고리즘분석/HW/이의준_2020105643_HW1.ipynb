```python
# algorithm A
def insertion_sort(arr):
    for end in range(1, len(arr)):
        for i in range(end, 0, -1):
            if arr[i - 1] > arr[i]:
                arr[i - 1], arr[i] = arr[i], arr[i - 1]

# algorithm B
def split(values, first, last):
        splitVal = values[(first+last)//2]
        while first <= last:
            while values[first] < splitVal:
                first += 1
            while values[last] > splitVal:
                last -= 1
            if first <= last:
                values[first], values[last] = values[last], values[first]   # swap
                first, last = first + 1, last - 1
        return first

def quick_sort(values, first, last):
    if first < last:
        splitPoint = split(values, first, last)
        quick_sort(values, first, splitPoint-1)
        quick_sort(values, splitPoint, last) 
       
    return values
```

```python
from random import randrange

N = [5000, 10000]

print("================== Insertion Sort ==================")
for n in N:
    print(f"n = {n}")
    print()
    print("[Before]")
    A = [randrange(1, 1001) for i in range(0, n)]
    print(A)
    insertion_sort(A)
    print("[After]")
    print(A)
    print()

print("================== Quick Sort ==================")
for n in N:
    print(f"n = {n}")
    print()
    print("[Before]")
    A = [randrange(1, 1001) for i in range(0, n)]
    print(A)
    quick_sort(A, 0, len(A)-1)
    print("[After]")
    print(A)
    print()
```

시간 측정

```python
import time 

N = [5000, 10000, 20000, 30000, 40000, 80000]

print("================== Insertion Sort ==================")
for n in N:
    if (n > 40000):
        break;
    
    print(f"n = {n}\n")
    A = [randrange(1, 1001) for i in range(0, n)]
    start = time.time()
    insertion_sort(A)
    end = time.time()
    print(f"{end - start:.6f} sec")
    print()

print("================== Quick Sort ==================")
for n in N:
    print(f"n = {n}\n")
    A = [randrange(1, 1001) for i in range(0, n)]
    start = time.time()
    quick_sort(A, 0, len(A)-1)
    end = time.time()
    print(f"{end - start:.6f} sec")
    print()
```
