```python
############################## Q1 #################################
def dijkstra(n, W, F):
    inf = 1000
    int_inf = 9999
    vnear = None
    touch = n*[0]   # touch[i]: v1부터 vi까지의 최단 거리 중 vi와 가장 가까운 v in Y
    length = n*[0]  # length[i]: v1부터 vi까지의 거리
    save_length = n*[0]

    for i in range(1, n):
        length[i] = W[0][i]
        touch[i] = 0    
    
    iter = n-1
    while(iter):
        min = int_inf
        for i in range(1,n):
            if (0 <= length[i] and length[i] < min):
                min = length[i]
                vnear = i

        e = (touch[vnear], vnear)
        save_length[vnear] = length[vnear]
        F.add(e)

        for i in range(1, n):
            if (length[vnear] + W[vnear][i] < length[i]):
                length[i] = length[vnear] + W[vnear][i]
                touch[i] = vnear

        length[vnear] = -1
        iter -= 1

    # Calculate NoC = Max Arc Num - v1-connected Acr Num
    NoC = 0
    max_arc_num = 0
    v1_connected_arc_num = 0
    for w_vi in W:
        for wi in w_vi:
            if wi != 0 and wi != inf:
                max_arc_num += 1

    for w in W[0]:
        if w != 0 and w != inf:
            v1_connected_arc_num += 1
    NoC = max_arc_num - v1_connected_arc_num

    return NoC, save_length
```

```python
inf = 1000
f = set()
w = [
     [0, 15, 4, 11, 5], 
     [inf, 0, inf, 1, inf], 
     [inf, 4, 0, 2, inf], 
     [inf, inf, inf, 0, inf], 
     [inf, 3, inf, 1, 0]
     ]
n = 5

NoC, save_length = dijkstra(n, w, f)

print('############################## Q1 #################################')
print(f)    # (3, 1), (0, 2), (4, 3), (0, 4)
print(save_length)
print(NoC)
print()
```

```python
############################## Q2 #################################
def promising(i, col):
    k = 0
    switch = True

    while (k<i  and switch):
        if col[i] == col[k] or abs(col[i] - col[k]) == i-k:
            switch = False
        k = k+1

    return switch

def queens(n, i, col, res):
    res[1] += 1     # 상태공간트리의 노드 수
    if promising(i, col):
        if (i == n-1):
            res[0] += 1     # 해의 개수
        else:
            for j in range(n):
                col[i+1] = j
                queens(n, i+1, col, res)

n = 7
col = n * [0]
res = [0, 0]   # call by reference를 위하여 list로 구현. res[0] = 해의 총 개수, res[1] = 모든 해를 찾기 위해 생성한 상태공간트리의 총 노드 수
queens(n, -1 , col, res)

print('############################## Q2 #################################')
# 상태공간트리의 노드 수는 start 노드를 제외하고 계산하였습니다.
print(f'모든 해의 개수: {res[0]}, 상태공간트리의 총 노드 수: {res[1]}')
```
