```python
def add_many(*args):
    result = []
    for i in args:
        result.append(str(i))
    return " ".join(result)

print(add_many('hi', 'hello'))
print(add_many(1,2,3,4))
```

# **File I/O**

```python
f = open('file.txt', 'w')

for i in range(10):
    data = "%d번째 줄입니다.\n" % i
    f.write(data)

f.close()
```

```python
f = open('file.txt', 'r')

while True:
    line = f.readline()
    if not line:
        break
    print(line, end="")

f.close()
```

```python
f = open('file.txt', 'r')
lines = f.readlines()

for line in lines:
    print(line, end="")

f.close()
```

```python
f = open('file.txt', 'r')
data = f.read()
print(data, end="")
f.close()
```
